```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# example: get historical data for apple(aapl)
ticker = yf.Ticker("AAPL")
#appl's data for the next 5 years
data=ticker.history(start="2020-01-01",end="2024-12-31",interval="1d")

#print or preview the data
print(data.head())
```

```python
plt.figure(figsize=(12,6))
plt.plot(data.index,data["Close"],label="Close Price")
plt.title("AAPL Closing Price")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()
plt.grid()
plt.show()
```
